```python
import nltk
from nltk.corpus import stopwords
from pymystem3 import Mystem
from string import punctuation


import numpy as np
import pandas as pd
```

```python
train = pd.read_excel('/content/Общий датасет.xlsx')
```

```python
for i in range(10):
    print(train.iloc[[i]]['Диалог'].values[0])
    print('______________________________')
```

```python
def take_user_part(x):
    i = 0
    user_array = []
    x_ = x.split()
    while i < len(x_):
        if x_[i] == 'К:' or x_[i] == '/nК:':
            i +=1
            user_array.append(x_[i].lower())
            while i < len(x_)-1 and x_[i+1] != '/nБ:' and x_[i+1] != 'Б:' :
                i += 1
                user_array.append(x_[i].lower())
        i +=1
    return ' '.join(user_array)
```

```python
def take_robot_part(x):
    i = 0
    user_array = []
    x_ = x.split()
    while i < len(x_):
        if x_[i] == 'Б:' or x_[i] == '/nБ:':
            i +=1
            user_array.append(x_[i].lower())
            while i < len(x_)-1 and x_[i+1] != '/nК:' and x_[i+1] != 'К:' :
                i += 1
                user_array.append(x_[i].lower())
        i +=1
    return ' '.join(user_array)
```

```python
nltk.download("stopwords")
mystem = Mystem()
russian_stopwords = stopwords.words("russian")

def preprocess_text(text):
    tokens = mystem.lemmatize(text.lower())
    tokens = [token for token in tokens if (token in ['не', 'нет', 'да', 'ни', 'нельзя', 'всегда', 'конечно'])\
              or (token not in russian_stopwords\
              and token != " " \
              and token.strip() not in punctuation)]

    text = " ".join(tokens)

    return text
```

```python
train['user_part'] = train['Диалог'].apply(take_user_part)
```

```python
train['robot_part'] = train['Диалог'].apply(take_robot_part)
```
